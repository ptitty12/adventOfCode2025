```python
ztups = []
for i in zinput.split(","):
    
    ztups.append([int(j) for j in i.split("-")])
        
```

```python
def checker(id):
    zlen = len(str(id)) // 2
    return str(id)[:zlen] == str(id)[zlen:]    
```

```python
runTot = 0
for k in ztups:
    for i in range(k[0],k[1]+1):
        if checker(i):
            runTot += i
runTot
```

```python
error = []
for k in ztups:
    for id in range(k[0],k[1]+1):
        for i in range(1,1+len(str(id))//2):

            check_for = str(id)[:i]
            multi = len(str(id))/i
            if int(multi) != multi:
                continue
            if check_for * (int(multi)-1) == str(id)[i:]:
                error.append(id)
                break
```
