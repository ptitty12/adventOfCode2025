```python
beams = []
all_beams = []
beams.append((zlist[0].find('S'),0))
i = 0
for step in range(0,len(zlist)-1):
    for x in range(0,len(zlist[step])):
        if zlist[step][x] == '^' and x in [beam[0] for beam in beams]:
            i+=1
   
            l = (x-1,step)
            r = (x+1,step)
            all_beams.append(l)
            all_beams.append(r)
            for b in [beam for beam in beams if beam[0] == x]:
                beams.remove(b)
            if l not in beams:
                beams.append(l)
            if r not in beams:
                beams.append(r)
```

```python
def next_spots(current):
    cx = current[0]
    cy = current[1]
    below = [beam for beam in all_beams if abs(beam[0] - cx) ==1 and beam[1] == cy+1]
    return below
```
