```python
def refresh(p):
    tbl2 = []

    
    for line in p.splitlines():
        tbl2.append(line)
    return tbl2
```

```python
def get_item(cords):

    x = cords[0]
    y = cords[1]
    if x <= -1 or y<= -1:
        return None
    try:
        return tbl2[y][x]
    except:
        return None
```

```python
def get_around(cords):
    # top left, top middle, then loop
    x= cords[0]
    y= cords[1]
    p1 = x-1, y-1
    p2 = x+0, y-1
    p3=x+1, y-1
    p4=x-1, y+0
    p5=x+1, y+0
    p6=x-1, y+1
    p7=x+0, y+1
    p8=x+1, y+1
    return [p1,p2,p3,p4,p5,p6,p7,p8]
    
```
